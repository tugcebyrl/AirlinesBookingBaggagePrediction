```python
df=pd.read_csv('/kaggle/input/airlines-booking-csv/customer_booking.csv',encoding="latin-1")
```

```python
df.head(3)
```

```python
df_new=pd.get_dummies(df,columns=["sales_channel","trip_type","flight_day","route","booking_origin"],drop_first=True)
```

```python
df_new.info()
```

```python
y=df_new["wants_extra_baggage"]
x=df_new.drop("wants_extra_baggage",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=35,train_size=0.85)
```

```python
rf=RandomForestClassifier()
model_rf=rf.fit(x_train,y_train)
```

```python
model_rf.score(x_test,y_test)
```

```python
crossVal=cross_val_score(model_rf,x,y,cv=4)
```

```python
print("Cross Val: ",crossVal)
```
